```python
# Dependencies and Set up
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
#connecting with the csv
df = pd.read_csv('Resources/crypto_data.csv')
df
```

```python
df.shape
```

```python
# filtering the data to only true values
df = df.loc[df['IsTrading']==True]
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df
```

```python
labels = df['CoinName']
labels
```

```python
df = df.drop(['IsTrading', 'CoinName'], axis=1)
df
```

```python
#resetting the index column 
df = df.set_index('Unnamed: 0')
df
```

```python
df.shape
```

```python
#filtering for values grater than 0
df = df[df['TotalCoinsMined'] > 0]
df
```

```python
df.shape
```

```python
#converting into dummies Algorithm and ProofType columns
X = pd.get_dummies(df, columns=['Algorithm', 'ProofType'])
X
```

```python
X.shape
```

```python
# Create a scaler to standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
# Train the scaler with the training data.
df_scaled = StandardScaler().fit_transform(X)
print(df_scaled)
```

```python
from sklearn.decomposition import PCA
```

```python
# Initialize PCA model
pca = PCA(n_components=0.99)

# Get two principal components for the iris data.
df_pca = pca.fit_transform(df_scaled)
df_pca
```

```python
df_pca.shape
```

```python
pca = PCA(n_components=3)

# Get two principal components for the iris data.
df_pca_cp3 = pca.fit_transform(df_scaled)
df_pca_cp3
```

```python
# Transform PCA data to a DataFrame
df_pca_cp3 = pd.DataFrame(
    data= df_pca_cp3, columns=["principal component 1", "principal component 2","principal component 3" ]
)
df_pca_cp3.head()
```

```python
#check what the values mean 
print(pca.explained_variance_)
```
